```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(6)
y1 = np.array([85.5,91,72,59,66,55])
y2 = np.array([94,82,89.5,62,49,53])
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
labels = ["语文","数学","英语","物理","化学","生物"]
bar_width = 0.4
plt.bar(x,y1,label='男',tick_label=labels,width=bar_width)
plt.bar(x+bar_width,y2,label='女',width=bar_width)
plt.legend(loc='upper right')
plt.xlabel("学科")
plt.ylabel("平均成绩")
plt.title('高二男生、女生各科的平均成绩柱形图')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(6)
y1 = np.array([85.5,91,72,59,66,55])
y2 = np.array([94,82,89.5,62,49,53])
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
labels = ["语文","数学","英语","物理","化学","生物"]
bar_width = 0.5
plt.bar(x,y2,label='女',tick_label=labels,bottom=y1,width=bar_width)
plt.bar(x,y1,label='男',width=bar_width)
plt.legend(loc='upper right')
plt.xlabel("学科")
plt.ylabel("平均成绩")
plt.title('高二男生、女生各科的平均成绩柱形图')
plt.show()
```
